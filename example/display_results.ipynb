```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
def cross_section_index(direction, cross, x, y, z):
    
    x0 = np.unique(x)
    y0 = np.unique(y)
    z0 = np.unique(z)

    idx = []
    if direction == 'EW' or direction == 'ew':
            # x(EW)-z cross section
        dy = y0[2] - y0[1]
        idx = np.where(np.abs(y - cross) < dy / 2.)[0]
    elif direction == 'NS' or direction == 'ns':
            # y(NS)-z cross section
        dx = x0[2] - x0[1]
        idx = np.where(np.abs(x - cross) < dx / 2.)[0]
    elif direction == 'H' or direction == 'h':
            # y(NS)-z cross section
        dz = z0[2] - z0[1]
        idx = np.where(np.abs(z - cross) < dz / 2.)[0]
    else:
            raise Exception('cross_section: direction invalid: EW(ew) or NS(ns) or H(h)')

    if len(idx) <= 0:
        raise Exception('cross_section: range invalid')

    return idx
```

```python
plt.rcParams['font.size'] = 12
```

```python
ret = np.loadtxt('synth/input_mag.in')
x0 = ret[:, 0]
y0 = ret[:, 1]
z0 = ret[:, 2]
f0 = ret[:, 3]

ret = np.loadtxt('synth/input_grv.in')
g0 = ret[:, 3]

nx0 = len(np.unique(x0))
ny0 = len(np.unique(y0))

print(nx0, ny0)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 2, 1)
pc = ax.pcolor(x0.reshape(nx0, ny0), y0.reshape(nx0, ny0), f0.reshape(nx0, ny0), cmap='jet')
ax.set_aspect('equal')
ax.set_title('Magnetic anomaly')
ax.set_xlabel('Y(km)')
ax.set_ylabel('X(km)')
plt.colorbar(pc, orientation='horizontal', shrink=0.8, pad=0.1, label='nT')

ax = fig.add_subplot(1, 2, 2)
pc = ax.pcolor(x0.reshape(nx0, ny0), y0.reshape(nx0, ny0), g0.reshape(nx0, ny0), cmap='jet')
ax.set_aspect('equal')
ax.set_title('Gravity anomaly')
ax.set_xlabel('Y(km)')
#ax.set_ylabel('X(km)')
plt.colorbar(pc, orientation='horizontal', shrink=0.8, pad=0.1, label='mGal')
```

```python
ret0 = np.loadtxt('synth/true_model.data')
x = ret0[:, 0]
y = ret0[:, 1]
z = ret0[:, 2]
b0 = ret0[:, 3]
p0 = ret0[:, 4]

nx = len(np.unique(x))
ny = len(np.unique(y))
nz = len(np.unique(z))

print(nx, ny, nz)
```

```python
fig = plt.figure(figsize=(12, 4))

# x(EW)-z cross section
yc = -0.05
idx = cross_section_index('EW', yc, x, y, z)

xx = x[idx].reshape(nz, nx)
zz = z[idx].reshape(nz, nx)
pp0 = p0[idx].reshape(nz, nx)
bb0 = b0[idx].reshape(nz, nx)

ax = fig.add_subplot(1, 2, 1)
ax.set_aspect('equal')
pc = ax.pcolor(xx, zz, bb0, cmap='jet')
pc.set_clim(-1, 1)
ax.set_ylabel('z(km)')
ax.set_xlabel('y(km)')
ax.set_title('True magnetization model')
plt.colorbar(pc, orientation='horizontal', label='A/m', shrink=0.5, pad=0.22)

ax = fig.add_subplot(1, 2, 2)
ax.set_aspect('equal')
pc = ax.pcolor(xx, zz, pp0, cmap='jet')
pc.set_clim(0, 1)
ax.set_ylabel('z(km)')
ax.set_xlabel('y(km)')
ax.set_title('True density model')
plt.colorbar(pc, orientation='horizontal', label='g/cm^3', shrink=0.5, pad=0.22)
```

```python
fig = plt.figure(figsize=(12, 4))

ret = np.loadtxt('res/model.data')
b = ret[:, 3]
p = ret[:, 4]

# x(EW)-z cross section
yc = -0.05
idx = cross_section_index('EW', yc, x, y, z)

xx = x[idx].reshape(nz, nx)
zz = z[idx].reshape(nz, nx)
pp = p[idx].reshape(nz, nx)
bb = b[idx].reshape(nz, nx)

ax = fig.add_subplot(1, 2, 1)
ax.set_aspect('equal')
pc = ax.pcolor(xx, zz, bb, cmap='jet')
pc.set_clim(-1, 1)
ax.set_ylabel('z(km)')
ax.set_xlabel('y(km)')
ax.set_title('Magnetization model')
plt.colorbar(pc, orientation='horizontal', label='A/m', shrink=0.5, pad=0.2)

ax = fig.add_subplot(1, 2, 2)
ax.set_aspect('equal')
pc = ax.pcolor(xx, zz, pp, cmap='jet')
pc.set_clim(0, 1.)
ax.set_ylabel('z(km)')
ax.set_xlabel('y(km)')
ax.set_title('Density model')
plt.colorbar(pc, orientation='horizontal', label='g/cm^3', shrink=0.5, pad=0.2)
```
